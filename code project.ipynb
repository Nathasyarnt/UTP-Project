```python
#panggil data
#import data
import pandas as pd
data = pd.read_excel("rawdata.xlsx")
print(data)
```

```python
# Identify non-numeric columns
non_numeric_columns = data.select_dtypes(exclude=['number']).columns
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Transform non-numeric columns
for column in non_numeric_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Display the updated dataframe
print(data)
```

```python
#Create a new column at the end of the dataframe that contains the same value 
data['Age_Years'] = data['Age']
#Remove the first column called age 
df = data.drop('Age', axis = 1)
#Show the dataframe
df
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X = data.iloc[:, 1:df.shape[1]].values 
Y = data.iloc[:, 0].values 
```

```python
# Split the dataset into 75% Training set and 25% Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#rqandom forest for classification
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(random_state=42)
forest.fit(X_train, Y_train)
forest.score(X_train, Y_train)

```
